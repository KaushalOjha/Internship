### 1. Scrape the details of most viewed videos on YouTube from Wikipedia.
### Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
### You need to find following details:
### A) Rank
### B) Name
### C) Artist
### D) Upload date
### E) Views

```python
from bs4 import BeautifulSoup
import selenium
from selenium import webdriver
import requests
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
# Run the driver
driver = webdriver.Chrome('chromedriver.exe')

# inserting url to driver
url1 = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url1)
```

```python
# to fetch and create the list of rank
Rank_ = driver.find_elements_by_xpath('/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr/td[1]')
print(len(Rank_))
rank = []
for i in Rank_:
    rank.append(i.text)

    
# to fetch and create the list of name    
Name_ = driver.find_elements_by_xpath('/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr/td[2]')
print(len(Name_))
name = []
for i in Name_:
    name.append(i.text)

    
# to fetch and create the list of artist
Artist_ = driver.find_elements_by_xpath('/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr/td[3]')
print(len(Artist_))
artist = []
for i in Artist_:
    artist.append(i.text)

    
# to fetch and create the list of upload date
Upload_Date_ = driver.find_elements_by_xpath('/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr/td[4]')
print(len(Upload_Date_))
Upload_Date = []
for i in Upload_Date_:
    Upload_Date.append(i.text)
    
    
# to fetch and create the list of Views    
Views_ = driver.find_elements_by_xpath('/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr/td[5]')
print(len(Views_))
Views = []
for i in Views_:
    Views.append(i.text)
```

```python
df1 = pd.DataFrame({"RANK":rank, "NAME":name, "ARTIST":artist, "UPLOAD DATE":Upload_Date, "VIEWS":Views})
df1
```

### 2. Scrape the details team India’s international fixtures from bcci.tv.
### Url = https://www.bcci.tv/.
### You need to find following details:
### A) Match title (I.e. 1st ODI)
### B) Series
### C) Place
### D) Date
### E) Time
### Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
# Run the driver
driver = webdriver.Chrome('chromedriver.exe')

# inserting url to driver
url1 = ' https://www.bcci.tv/'
driver.get(url1)
```

```python
# To reach at international fixture page
ALL_fixture = driver.find_element_by_xpath('/html/body/div[5]/div/div[2]/div/div[2]/div/div[1]/div[2]/a')
ALL_fixture.click()
```

```python
# to fetch and create the list of matches
Matches_ = driver.find_elements_by_xpath('//span[@class="ng-binding"]')
print(len(Matches_))
Matches = []
for i in Matches_:
    Matches.append(i.text)

# to fetch and create the list of series 
Series_ = driver.find_elements_by_xpath('//span[@class="ng-binding ng-scope"]')
print(len(Series_))
Series = []
for i in Series_:
    Series.append(i.text)
    
# to fetch and create the list of place
Place_ = driver.find_elements_by_xpath('//div[@class="fix-place ng-binding ng-scope"]')
print(len(Place_))
Place = []
for i in Place_:
    Place.append(i.text)
    
# to fetch and create the list of date
Date_ = driver.find_elements_by_xpath('//h5[@class="ng-binding"]')
print(len(Date_))
Date = []
for i in Date_:
    Date.append(i.text)

# to fetch and create the list of date
Time_ = driver.find_elements_by_xpath('//h5[@class="text-right ng-binding"]')
print(len(Time_))
Time = []
for i in Time_:
    Time.append(i.text)
```

```python
df2 = pd.DataFrame({"MATCHES":Matches, "SERIES":Series, "PLACE":Place, "DATE":Date, "TIME":Time})
df2
```

```python
df3 = pd.DataFrame({"NAMES":ExName, "Description":ExDesc })
df3
```
